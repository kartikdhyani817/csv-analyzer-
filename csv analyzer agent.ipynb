```python
import pandas as pd

df = pd.read_csv(r"C:\Users\HP\Downloads\high_sales_days.csv")
df.head()
```

```python
def dataframe_to_text(df, max_rows=5):
    """Convert first few rows of a DataFrame to readable text."""
    preview = df.head(max_rows)
    return preview.to_string(index=False)

context = dataframe_to_text(df)
print("Context Preview:\n", context)
```

```python
import requests

def ask_ollama_about_csv(prompt, context, model="mistral"):
    full_prompt = f"""You are a data assistant. Given this CSV preview:

{context}

Answer this question: {prompt}
"""
    response = requests.post(
        "http://localhost:11434/api/generate",
        json={
            "model": model,
            "prompt": full_prompt,
            "stream": False
        }
    )
    result = response.json()
    return result.get("response", result.get("error", "Unknown error"))
```

```python
question = "Which column has the highest average?"
answer = ask_ollama_about_csv(question, context)
print("Answer:", answer)
```
